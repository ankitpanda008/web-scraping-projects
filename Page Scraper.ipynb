```python
data = []

for sp in soup.find_all('div', class_='quote'):
    quote = sp.find('span', class_ = 'text').text[1:-1]
    author = sp.find('small').text
    author_id = sp.find('a').get('href')
    tags = []
    for i in sp.find_all('a', class_='tag'):
        tags.append(i.text)
    
    tags = ','.join(tags)
    data.append([quote, author, author_id, tags])
    
    
  
# First save the author name and quote and then use the get to collect the href.
```

## Scraping the data from multiple pages and how to automate the process.

```python
data = []
link = 'https://quotes.toscrape.com/page/'
for page in tqdm(range(1,11)):
    link_main = link +str(page)
    res = requests.get(link_main)
    soup = BeautifulSoup(res.text, 'html.parser')
    
#     for sp in soup.find_all('div', class_='quote'):

    for sp in soup.find_all('div', class_='quote'):
        quote = sp.find('span', class_ = 'text').text[1:-1]
        author = sp.find('small').text
        author_id = sp.find('a').get('href')
        tags = []
        for i in sp.find_all('a', class_='tag'):
            tags.append(i.text)
    
        tags = ','.join(tags)
        data.append([quote, author, author_id, tags])
```

```python
len(data)
```

```python
df = pd.DataFrame(data, columns=['Quote', 'author', 'author_id', 'tags'])
```

```python
df.to_csv('Quotes.csv', index=False)
```

***Task to perform:***
*To get into each link for author and save their data in a separate text file.*

```python
author_info = []
paths = []
data = []
for author in tqdm(df.values):
    name = 'Authors/' + author[0] + '.txt'
    link = author[1]
    
    res = requests.get(link)
    soup = BeautifulSoup(res.text, 'html.parser')
    text = soup.find('div', class_='author-details').text
    
    
    author_info.append(text.strip())
    paths.append(name)
    fd = open(name, 'w', encoding='utf-8')
    fd.write(text)
    fd.close()
    # One thing done before executing is to see if there are more than one quotes by single authors and removing the duplicates.
    
```

```python
# We can make a csv file for authors as well.

df['author_info'] = author_info
df['paths'] = paths
```
