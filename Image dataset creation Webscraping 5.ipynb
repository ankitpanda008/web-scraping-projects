### 1.Task: carry out the operation of fetching the unique tags so we can group them into folders later.

```python
# USing list comprehension to get rid of the leading spaces in tags

t = []
for tags in df['tags']:
    t += [tag.strip() for tag in tags.split(',')]
    
print(len(t))
print(len(set(t)))

tags = list(set(t))
tags
```

### Creating folders for all the tags

```python
tags = tags[1:]
# print(tags)
for tag in tqdm(tags):
    try:
        os.mkdir('Image-stock-tags-folder/'+ tag)
    except:
        continue
        
```
